# 1.3 Accesing the Contents of the Data.CSV file

```python
npz_paths = []

IMAGE_RESIZE = (256,256)
# merged_df = merged_df.reset_index(drop=True)

for i, row in merged_df.iterrows():
    picture_Path = row['Path']

    npz_path = picture_Path.split('.')[0] + '.npz'
    npz_paths.append(npz_path)


    pic_bgr_arr = cv2.imread(picture_Path)
    pic_rgb_arr = cv2.cvtColor(pic_bgr_arr, cv2.COLOR_BGR2RGB)
    pic_rgb_arr = cv2.resize(pic_rgb_arr, IMAGE_RESIZE)

    artist = np.array(row['Artist'])

    lowEst = np.array(row['LowEstimationPrice']).astype('int')
    highEst = np.array(row['HighEstimationPrice']).astype('int')

    meanEst = int ((lowEst + highEst) / 2)
    hammerPrice = np.array(row['HammerPrice'])
    

    estimations = np.array([lowEst, highEst, meanEst])
    estimations.astype('int')
    
    np.savez_compressed(npz_path, pic=pic_rgb_arr, artist=artist, estimations=estimations, hammerPrice=hammerPrice)
    
merged_df['NPZ_Path'] = pd.Series(npz_paths)


    

print(merged_df.head())
print(merged_df.tail())
```

```python
features_df  = merged_df[['Artist','LowEstimationPrice', 'HighEstimationPrice', 'HammerPrice']]

# features_df["ID"] = features_df['ID'].astype(int)

features_df.head()
```

```python

maxLow = features_df['LowEstimationPrice'].max()
maxHigh = features_df['HighEstimationPrice'].max()
maxSale = features_df['HammerPrice'].max()
# print(maxLow)
# print(maxHigh)
# print(maxSale)
means = []

for i, row in features_df.iterrows():
    features_df.loc[i, 'LowEstimationPrice'] = features_df.at[i, 'LowEstimationPrice'] / maxLow
    # row['HighEstimationPrice']  = row['HighEstimationPrice'] / maxHigh
    features_df.loc[i, 'HighEstimationPrice'] = features_df.at[i, 'HighEstimationPrice'] / maxHigh
    mean = (features_df.at[i, 'HighEstimationPrice'] + features_df.at[i, 'LowEstimationPrice'] ) / 2
    means.append(mean)
    features_df.loc[i, 'HammerPrice'] = features_df.at[i, 'HammerPrice'] / maxSale

features_df['MeanEstimation'] = pd.Series(means)
features_df.head()

```

# 3 - Loading in the Data

```python


def get_X_y(df):
    X_pic, X_artist, X_estimations = [], [], []
    y = []

    for name in df['NPZ_Path']: 
        loaded_npz = np.load(name)

        pic = loaded_npz['pic']
        X_pic.append(pic)

        estimations = loaded_npz['estimations']
        X_estimations.append(estimations)

        artit = loaded_npz['artist']
        # print(artit)

        X_artist.append(artit)

        sale = loaded_npz['hammerPrice']
        y.append(sale)

    X_pic, X_artist, X_estimations = np.array(X_pic), np.array(X_artist), np.array(X_estimations)
    y = np.array(y)

    return (X_pic, X_artist, X_estimations), y

```

```python
(X_train_pic, X_train_artist, X_train_estimations), y_train = get_X_y(train_df)

(X_val_pic, X_val_artist, X_val_estimations), y_val = get_X_y(val_df)

(X_test_pic, X_test_artist, X_test_estimations), y_test = get_X_y(test_df)
```
